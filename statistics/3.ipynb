# 3.통계적 실험과 유의성검정
실험설계는 어떤 가설을 확인하거나 기각하기 위한 목표로서 통계분석의 토대가 된다.   
통계적 유익성, t 검정 p 값 등은 전형적인 통계적 추론이라는 파이프라인 속에 있음을 알 수 있다. 이는 A가 B보다 낫다, A가 수익성이 높다 등의 가설을 세우는 것으로 출발한다. 여러 실험은 가설을 검정하고 원하는 최종적 결론을 도출할 수 있도록 설계된다. 그리고 데이터를 수집하고 분석한 뒤 결론을 도출한다.   
추론inference이라는 용어는 제한된 데이터로 주어진 실험 결과를 더 큰 과정 혹은 모집단에 적용하려는 의도를 반영한다.


## 3-1.A/B 검정
A/B 검정은 두 가지 방식(처리 방법, 제품, 절차) 중 어느 쪽이 다른 쪽보다 더 우월하다는 것을 입증하기 위해 실험군을 두 그룹으로 나눠 진행하는 실험이다.   
두 처리 방법 중 하나는 기준이 되는 기존 방법이거나 아무런 처리도 적용하지 않는 방법이 된다. 이를 대조군이라고 한다. 주로 새로운 처리 방법을 적용하는 것이 대조군보다 낫다는 것이 일반적인 가설이 된다. 

- 처리: 어떤 대상에 주어지는 특별한 환경이나 조건
- 처리군: 특정 처리에 노출된 대상들의 집단 (대조군의 반대)
- randomization: 처리를 적용할 대상을 임의로 결정하는 과정
- 대상: 처리를 적용할 개체 대상
- 검정통계량: 처리 효과를 측정하기 위한 지표 (그룹 A와 B를 비교하는 데 사용)

A/B 검정에는 둘 중 어느 한쪽의 처리를 할당할 수 있는 대상이 주어지고, 대상은 무작위로 특정 처리에 노출된다.이때 처리군 간의 차이는 아래 두 가지 이유 중 하나 때문이라고 할 수 있다. 
- 다른 처리의 효과
- 어떤 대상이 어떤 처리에 배정될지에 대한 경우의 수(즉, 무작위로 배정한 결과가 더 좋은 결과를 보이는 한쪽에 집중됨)

또한 검정통계량 또는 측정 지표에 주의를 기울여야 한다. 

## 3-2.가설검정
가설검정 혹은 유의성검정의 목적은 관찰된 효과가 우연에 의한 것인지 여부를 알아내는 것이다.   

- 귀무가설: 우연 때문이라는 가설
- 대립가설: 귀무가설과의 대조(증명하고자 하는 가설)
- 일원검정: 한 반향으로만 우연히 일어날 확률을 계산하는 가설검정
- 이원검정: 양방향으로 우연히 일어날 확률을 계산하느 가설검정

### 3-2-1.귀무가설
가설검정은 다음과 같은 논리를 사용한다. '인간은 실제로 우연히 발생한 일이라도 그것이 흔하지 않다면, 뭔가 의미가 있을 거라고 해석하는 경향이 있다. 그러므로 실험에서 얻은 그룹 간의 차이가 무작위로 얻을 수 있는 합리적인 수준과는 다르다는 증거가 필요하다.'  
또한 그룹들이 보이는 결과는 서로 동일하며, 그룹 간의 차이는 우연에 의한 결과라는 것을 기본 가정으로 설정한다. 이 기본 가정을 귀무가설이라고 부른다.   
결국 가설이 틀렸다는 것을 입증해서 두 그룹 간의 차이가 우연이 아니라는 것을 보여줘야하는 것이다. 

### 3-2-2.대립가설
가설검정은 귀무가설뿐만 아니라 그와 대립하는 가설을 포함한다. 
- 귀무가설: 두 집단 간의 평균에는 차이가 없을 것이다.  
대립가설: 두 집단 간의 평균은 차이가 있을 것이다. 
- 귀무가설: A$\leq$B  
대립가설: A>B
- 귀무가설: B는 A보다 x% 더 작다.   
대립가설: B는 A보다 x% 더 크다.

### 3-2-3.일원/이원 가설검정
- 일원 가설검정: 단방향 대립가설이 필요한 경우 사용, 우연에 의한 극단적인 결과에 대해 한 방향만을 고려하여 p값을 계산한다는 의미. 
- 이원 가설검정: 양방향 대립가설이 필요한 경우 사용, 우연에 의한 극단적인 결과가 양쪽에서 나타날 p 값을 계산한다는 것을 의미.

새로운 그룹이 더 좋은 것으로 증명되지 않는 한 일반적으로 원래 그룹이 기본값으로 지정되는 상황에서는 의사 결정을 필요로하는 A/B 검정의 특성상 단방향과 잘 어울린다고 할 수 있다. 

## 3-3.재표본추출
통계학에서 재표본추출은 랜덤한 변동성을 알아보자는 일반적인 목표를 가지고, 관찰된 데이터의 값에서 표본을 반복적으로 추출하는 것을 의미한다.   
재표본추출에는 부트스트랩과 순열검정이라는 주요 유형이 있다.

### 3-3-1.순열검정
두 개 이상의 표본을 사용하며, 일반적으로 A/B 또는 기타 가설검정을 위해 사용한다.

순열검정의 과정은 아래와 같다.
1. 여러 그룹의 결과를 하나의 데이터 집합으로 합친다.
2. 결합된 데이터를 섞은 뒤 ,그룹 A와 동일한 크기의 표복을 무작위로 비복원 추출한다.
3. 나머지 그룹에서도 그 그룹과 동일한 크기의 샘플을 무작위로 비복원 추출한다.
4. 새로 추출한 재표본에 대해 다시 통계량 혹은 추정값을 구하고 기록한다. (이것으로 한 번의 순열 반복이 진행된다.)
5. 앞선 단계들을 R번 반복하여 검정통계량의 순열분포를 얻는다.

실험을 통해 관찰한 그룹 간의 차이와 순열 과정에서 얻은 집합에서의 차이를 도표화해 비교한다.

관찰한 차이가 순열로 보이는 차이의 집합 안에 들어 있다면,어떤 것도 증명할 수 없다. 즉, 관찰된 차이가 우연히 일어날 수 있는 범위 안에 있다는 말이다.

관찰한 차이가 대부분 순열 분포의 바깥쪽에 있다면, 우연 떄문이 아니라고 결론 내릴 수 있다. 이 차이는 **통계적으로 유의미**하다고 한다.

### 3-3-2.부트스트랩 순열검정
위의 랜덤 셔플링 절차를 임의순열검정 또는 임의화검정이라고 부르며, 이외에도 전체순열검정, 부트스트랩 순열검정이 있다.
- 전체순열검정에서는 데이터를 무작위로 섞고 나누는 대신 나눌 수 있는 모든 가능한 조합을 찾는다. 이는 샘플 크기가 비교적 작을 때만 실용적인데, 셔플링을 반복할수록, 임의순열검정 결과는 전체순열검정의 결과와 유사하게 근접한다.
- 부트스트랩 순열검정에서는 무작위 순열검정의 2,3단계에서 비복원 대신 복원추출로 수행한다. 이로써 재표본추출 과정에서 모집단에서 개체를 선택할 때와 개체가 처리군에 할당될 때 임의성을 보장한다.b

## 3-4.통계적 유의성과 p값
통계적 유의성이란, 실험 결과가 우연히 일어난 것인지 아니면 우연히 일어날 수 없는 것인지를 판단하는 방법이다. 결과가 우연히 벌어질 수 있는 변동성의 밖에 있다면 이는 통계적으로 유의미하다.

- p 값: 귀무가설을 전제로, 관측된 결과와 같이 특이하거나 극단적인 결과를 얻을 확률
- 유의수준(alpha): 실제 결과가 통계적으로 의미 있는 것으로 간주되기 위해, 우연에 의한 결과가 능가해야 하는 비정상적인 가능성의 임계 확률 (5%, 1%가 주로 쓰임)
- 제1종 오류: 우연에 의한 효과를 실제 효과라고 잘못 결론 내리는 것
- 제2종 오류: 실제 효과를 우연에 의한 효과라고 잘못 결론 내리는 것

### 3-4-1.p 값
순열검정으로 얻은 결과 중, 관찰된 차이와 같거나 더 큰 차이를 보이는 경우의 비율로 p값을 추정할 수 있다.

```python
import numpy as np
from scipy import stats

data = np.array([[200, 23739 - 200], [182, 22588 - 182]])
chi2, p_value, df, _ = stats.chi2_contingency(data)

print(f'p-value for single sided test: {p_value / 2:.4f}')
```

p 값은 0.308이다. 즉, 우연히 얻은 결과의 30% 정도가 관찰한 것과 비슷한 정도로 예외적인 결과를 얻을 것으로 기대할 수 있다.

사람들은 p 값을 통해 실제 의미하는 것보다 더 많은 의미를 찾으려 한다. 사실 p 값을 통해 전달하고자 하는 의미는 '결과가 우연에서 비롯될 확률'인데 말이다.
우리는 더 낮은 p 값을 원하고 결국 뭔가를 증명했다고 결론 내리기를 원하며 p 값을 이렇게 해석한다. 하지만 실제 p 값이 나타내는 것은 '랜덤 모델이 주어졌을 때, 그 결과가 관찰된 결과보다 더 극단적일 확률'이다.

### 3-4-3.제1종과 제2종 오류
통계적 유의성 평가시 1종 오류와 2종 오류가 발생할 수 있다.
2종 오류는 어떤 오류라기보단 표본크기가 너무 작아서 효과를 알아낼 수 없다고 판단하는 것과 같다. p 값이 통계적 유의성에 미치지 못하는 경우, 실제 의미는 '효과가 아직 입증되지 않았다'는 뜻이다. 표본크기가 클수록 p 값은 작아진다.
유의성검정의 기본 기능은 어쩌다 우연히 일어난 일에 속지 않도록 하는 것이다. 보통은 1종 오류를 최소화하도록 가설을 설계한다.

### 3-4-4.데이터 과학에서의 p값

데이터과학자에게 p 값은 유용한 모델의 결과가 일반적인 랜덤 변이의 범위 내에 있는지를 알고 싶을 때 유용한 측정 지표이다. p 값을 중간 입력으로 쓰기도, p 값에 따라 피처의 포함여부를 정하기도 한다.

## 3-5.t 검정
단일 표본평균의 분포를 근사화하기 위해 개발됨
- 검정통계량: 관심의 차이 또는 효과에 대한 측정 지표
- t 통계량: 평균과 같이 표준화된 형태의 일반적인 검정통계량
- t 분포: 관측된 t 통계량을 비교할 수 있는 기준분포

유의성검정은 관심 있는 효과를 측정하기 위한 검정통계량을 지정하고, 관찰된 효과가 정상적인 랜덤 변이의 범위 내에 있는지 여부를 판단하는 데 도움을 준다.   
 
- 컴퓨터 보급 전에는 재표본 검정은 실용적이지 않았고, 통계학자들은 대신 표준적인 분포를 참고했다. 이 떄문에 검정통계량이 표준화되어 참고할 분포와 비교할 수 있게 되었다.   
널리 사용되는 표준화된 통계량 중 하나가 t 통계량이다. 

## 3-6.다중검정
예를 들어 20개의 예측변수와 1개의 결과변수가 모두 임의로 생성되었을 때 유의수준 0.05에서 20번의 일련의 유의성검정을 수행하면 실수로라도 하나의 예측변수에서 통계적으로 유의미한 결과를 초래할 수도 있다.(=1종 오류)  
이 확률은 1에서 0.05(즉, 0.05의 유의수준에서 항상 유의미하지는 않다는 올바른 검정 결과가 나올 확률)를 빼서 구할 수 있다.  
무의미 하다고 정확하게 검정할 확률이 0.95이므로 20번 모두 무의미하다고 올바른 검정 결과를 보일 확률을 0.95^20 = 0.36이다.   
따라서 적어나 하나의 예측값이 유의미하다고 검정 결과가 나올 확률은 1 - 0.36(모든 것이 무의미하다는 결론이 나ㅏ올 확률) = 0.64이다.(알파 인플레이션)

| overfitting도 이런식으로 변수와, 모델이 많을 수록 우연에 의해 유의미한 것으로 나타날 확률이 커진다.

- 거짓 발견 비율FDR: 다중검정에서 1종 오류가 발생하는 비율
- 알파 인플레이션: 1종 오류를 만들 확률인 알파가 테스트를 수행할수록 증가하는 현상

연구 조사, 데이터 마이닝에서 다중성은 일부가 우연히 유의미하다는 결론을 내릴 위험성을 증가시킨다.  
여러 통계 비교와 관련된 상황의 경우, 통계적 수정 절차가 필요하다.

## 3-7.자유도
표본 데이터에서 계산된 통계량에 적용되며 변화가 가능한 값들의 개수  
예를 들어 10개의 값으로 이뤄진 표본에서 평균을 알고 있다면 9개의 자유도가 있는 것이다. (이는 표본 중 9개를 알고있다면, 평균을 알고있으므로 10번째는 계산할 수 있지만 자유롭게 변경할 수 없으므로)

- 자유도(d.f)는 검정통계량을 표준화하는 계산의 일부이며, 이를 통해 기준분포와 비교할 수 있다.
- 자유도 개념은 회귀를 할 때 범주형 변수들을 n-1지표 혹은 더미 변수로 요인화하는 것의 이유가 된다. 

## 3-8.분산분석
여러 그룹 A/B/C/D의 수치 데이터들을 서로 비교한다고 가정하면 여러 그룹 간 통계적으로 유의미한 차이를 검정하는 통계적 절차를 분산분석ANOVA라고 한다. 
- 쌍벌 비교: 여러 그룹 중 두 그룹 간의 가설검정

- 분산분해: 구성 요소 분리, 예를 들어 전체 평균, 처리 평균, 잔차 오차로부터 개별 값들에 대한 기여를 뜻함
