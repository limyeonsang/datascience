### OOP_style_plotting

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([5,10,15], float)
y = np.array([10,20,15], float)
```

```python
hf = plt.figure()
ha = hf.add_axes([0,0,1,1])
ha.plot(x,y)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x,y)
```

```python
x1 = np.array([10,11,12])
y1 = np.array([10,10,10])
line.set_data(x1, y1)
fig
```

```python
x = np.linspace(0,2,100)
y1 = 0.5*x
y2 = 0.5*x**2
y3 = 0.5*x**3
```

```python
# PyPlot API Style Plotting

plt.plot(x,y1, label='linear')
plt.plot(x,y2, label='quadratic')
plt.plot(x,y3, label='cubic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
# OOP Style Plotting

hf, ha = plt.subplots()
hp1, = ha.plot(x,y1)
hp2, = ha.plot(x,y2)
hp3, = ha.plot(x,y3)
ha.set_xlabel('x')
ha.set_ylabel('y')
hp1.set_label('linear')
hp2.set_label('quadratic')
hp3.set_label('cubic')
ha.legend()
```

```python
hp1.set_color('k')
hp2.set_color('b')
hp3.set_color('r')
ha.grid()
ha.legend()
hf
```
